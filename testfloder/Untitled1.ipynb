```python
import tensorflow as tf
import math
import numpy as np
sess = tf.InteractiveSession()
```

```python
e = np.random.randn(1000)
t = np.linspace(0.,1.,1000)
f = 4
u = np.sin(2 * math.pi * f * t)
x = tf.placeholder(dtype = tf.float32, shape = (1000))
y = x + 0.2*e
```

```python
import matplotlib.pyplot as plt
%matplotlib inline 
plt.plot(t, y.eval(session = sess, feed_dict = {x:u}))
```

```python
h = tf.fft[y]
```
